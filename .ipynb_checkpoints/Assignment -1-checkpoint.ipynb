```python
#Helper function to print VIF value
def getvif(data):
    #Creating a Dataframe for VIF values
    vif = pd.DataFrame()
    #Dropping const column
    temp = data.drop("const",axis=1)
    #Copying feature names
    vif['Features'] = temp.columns
    
    #CAlculating VIF
    vif['VIF'] = [variance_inflation_factor(temp.values, i) for i in range(temp.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    
    #Sorting dataframe in descending order based on VIF values
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
```

## Now lets try ridge and lasso techniques and see if we can do any better in terms of performance

# Ridge Regression

```python
#Creating a copy of train and test data frame
df_train_ridge=df_train.copy()
df_test_ridge=df_test.copy()

#Creating x and y of training set
y_train=df_train_ridge.pop("SalePrice")
X_train=df_train_ridge

#Creating x and y of training set
y_test=df_test_ridge.pop("SalePrice")
X_test=df_test_ridge
```

```python
# list of alphas to tune
params = {'alpha': [0.0001, 0.001, 0.01, 0.05, 0.1, 0.2,0.4, 0.5, 0.6,0.8, 0.9, 1.0, 2.0,
                    4.0, 5.0, 6.0, 8.0, 9.0, 10.0, 20, 50, 100,300,700,1000,2000]}

ridge = Ridge()

# cross validation
folds = 5
model_cv = GridSearchCV(estimator = ridge, 
                        param_grid = params, 
                        scoring= 'r2', 
                        cv = folds, 
                        return_train_score=True,
                        verbose = 1)            
model_cv.fit(X_train, y_train) 
```

```python
#Looking at best alpha
model_cv.best_params_
```

```python
#Checkingout metrics
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results.head()
```
